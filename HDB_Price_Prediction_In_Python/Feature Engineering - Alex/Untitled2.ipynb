```python
from math import radians, sin, cos, sqrt, atan2
from math import radians, cos, sin, asin, sqrt
from scipy.spatial import cKDTree
def haversine_cbd(row):
    # Radius of Earth in kilometers (use 3956 for miles)
    R = 6371.0
    lat1 = row["LATITUDE"]
    lon1 = row["LONGITUDE"]
    ## Raffles Place Lat and Long
    lat2 = 1.2830
    lon2 = 103.8513
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Distance in kilometers
    distance = R * c
    return distance
```

# This portion of the code is to determine the straightline distance to the nearest MRT

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in km
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    a = np.sin(delta_phi / 2)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    distance = R * c  # Distance in km
    return distance
```

```python
# Load the CSV files containing the house and mrt data
mrt = pd.read_csv('mrt.csv')  # Replace with your actual file path


# Convert latitude and longitude to numpy arrays
house_lats = df['LATITUDE'].values
house_lons = df['LONGITUDE'].values
mrt_lats = mrt['Latitude'].values
mrt_lons = mrt['Longitude'].values

# Compute the distances between all houses and supermarkets using broadcasting
# We need to calculate distances for every pair of house and supermarket
distances_matrix = haversine(
    house_lats[:, np.newaxis], house_lons[:, np.newaxis], 
    mrt_lats, mrt_lons
)

# Find the index of the closest supermarket for each house
closest_indices = np.argmin(distances_matrix, axis=1)

# Get the closest distances for each house
closest_distances = distances_matrix[np.arange(len(df)), closest_indices]

# Add the closest distance to the houses DataFrame
df['Closest_mrt_Distance'] = closest_distances

# Add the name or identifier of the closest MRT station to the houses DataFrame
df['Closest_mrt_Station'] = mrt['Name'].values[closest_indices]
```

```python
# Load the CSV files containing the house and busstop data
busstop = pd.read_csv('busstops.csv')  # Replace with your actual file path


# Convert latitude and longitude to numpy arrays
house_lats = df['LATITUDE'].values
house_lons = df['LONGITUDE'].values
busstop_lats = busstop['Latitude'].values
busstop_lons = busstop['Longitude'].values

# Compute the distances between all houses and supermarkets using broadcasting
# We need to calculate distances for every pair of house and supermarket
# Create a 2D array of school coordinates for KDTree
bus_coors = np.column_stack((busstop_lats, busstop_lons))

# Build a KDTree for schools to efficiently perform range queries
bus_tree = cKDTree(bus_coors)

# Define the distance threshold (1 km)
threshold_distance = 0.5  # in kilometers

# For each house, count how many schools are within 1 km using KDTree
busstop_density = []

for i in range(len(df)):
    house_coord = (house_lats[i], house_lons[i])  # Coordinates of the current house
    
    # Query the KDTree for all schools within the threshold distance
    indices_within_1km = bus_tree.query_ball_point(house_coord, threshold_distance/111.32)
    
    # Append the number of schools within 1 km
    busstop_density.append(len(indices_within_1km))
print(len(busstop_density))
# Add the school density as a new column to the houses DataFrame
df['busstop_Density_Within_1KM'] = busstop_density

# Now df contains the school density within 1km for each house
```
